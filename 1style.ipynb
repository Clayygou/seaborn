```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def sinplot(flip = 1 ):
    x = np.linspace(0,14,100)
    for i in range(1,7):
        plt.plot(x,np.sin(x + i *5) * (7-i)*flip)
```

```python
sinplot()
```

```python
import seaborn as sns
sns.set()  # 
sinplot()
```
