```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
iris = sns.load_dataset("iris")
sns.pairplot(iris)              # 数据之间两两相关的特征
```

```python
sns.set(color_codes = True)
tips = sns.load_dataset("tips")
tips.head()
```

```python
sns.regplot(x = "total_bill",y= "tip",data = tips)
```

```python
sns.lmplot(x = "size",y= "tip",data = tips)    # 不适合做回归分析
```

```python
sns.lmplot(x = "size",y= "tip",data = tips,x_jitter = .05) # 加上抖动
```

```python
sns.stripplot(x= "day",y= "total_bill",data = tips) # 默认jitter = True
```

```python
sns.swarmplot(x= "day",y= "total_bill",data = tips, hue = "sex")  # 不同颜色
```

# 盒图
 * IOR 是统计学的四分位距 
 * 离群点 是 [Q1 - 1.5IQR  ，Q3 + 1.5IQR] 之外的数
 * 上下两条线是最大最小值  盒子上下线是1/4 和3/4值 中间是中位数
 * 菱形是离群点

```python
sns.boxplot(x= "day",y= "total_bill",data = tips, hue = "time")
```

```python
sns.boxplot(data = iris,orient = "h") # h is 横着画图
```

```python
sns.violinplot(y= "day",x= "total_bill",data = tips, hue = "sex",split = True)
```

```python
sns.violinplot(x= "day",y= "total_bill",data = tips,inner = None)
sns.swarmplot(x = "day",y = "total_bill",data = tips,color = "w",alpha = .5) # 两张图的合并
```

```python
titanic = sns.load_dataset("titanic")
sns.barplot(x= "sex",y = "survived",hue = "class",data = titanic)  # 条形图
```

```python
sns.pointplot(x= "sex",y = "survived",hue = "class",data = titanic)  # 点图 差异性明显
```

```python
sns.pointplot(x= "class",y = "survived",hue ="sex",data = titanic,
             palette = {"male":"g","female":"m"},
             markers = ["^",","],linestyle = ["-","——"]) # how to hua xuxian?
```

```python
sns.factorplot(x = "day",y = "total_bill",hue = "smoker",data = tips) #baseline
```

```python
sns.factorplot(x = "day",y = "total_bill",hue = "smoker",data = tips, kind = "bar") #kind
```

```python
sns.factorplot(x = "day",y = "total_bill",hue = "smoker",data = tips,kind = "swarm",col = "time") #
```

```python
sns.factorplot(x = "time",y = "total_bill",hue = "smoker",data = tips,
              col = "day",kind = "box",size =4,aspect = .5) #
```
