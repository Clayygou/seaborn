```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
iris = sns.load_dataset("iris")
sns.pairplot(iris)              # 
```

```python
sns.set(color_codes = True)
tips = sns.load_dataset("tips")
tips.head()
```

```python
sns.regplot(x = "total_bill",y= "tip",data = tips)
```

```python
sns.lmplot(x = "size",y= "tip",data = tips)    # 
```

```python
sns.lmplot(x = "size",y= "tip",data = tips,x_jitter = .05) # 
```

```python
sns.stripplot(x= "day",y= "total_bill",data = tips) # jitter = True
```

```python
sns.swarmplot(x= "day",y= "total_bill",data = tips, hue = "sex")  # 
```

```python
sns.boxplot(x= "day",y= "total_bill",data = tips, hue = "time")
```

```python
sns.violinplot(y= "day",x= "total_bill",data = tips, hue = "sex",split = True)
```

```python
sns.violinplot(x= "day",y= "total_bill",data = tips,inner = None)
sns.swarmplot(x = "day",y = "total_bill",data = tips,color = "w",alpha = .5) # 
```

```python
titanic = sns.load_dataset("titanic")
sns.barplot(x= "sex",y = "survived",hue = "class",data = titanic)  # 
```

```python
sns.pointplot(x= "sex",y = "survived",hue = "class",data = titanic)  #  
```

```python
sns.pointplot(x= "class",y = "survived",hue ="sex",data = titanic,
             palette = {"male":"g","female":"m"},
             markers = ["^",","],linestyle = ["-",""]) # how to hua xuxian?
```

```python
sns.factorplot(x = "day",y = "total_bill",hue = "smoker",data = tips) #baseline
```

```python
sns.factorplot(x = "day",y = "total_bill",hue = "smoker",data = tips, kind = "bar") #kind
```

```python
sns.factorplot(x = "day",y = "total_bill",hue = "smoker",data = tips,kind = "swarm",col = "time") #
```

```python
sns.factorplot(x = "time",y = "total_bill",hue = "smoker",data = tips,
              col = "day",kind = "box",size =4,aspect = .5) #
```
