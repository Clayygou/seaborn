```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
sns.set(color_codes = True)
np.random.seed(sum(map(ord,"distributions")))
```

```python
x = np.random.normal(size = 100)
sns.distplot(x,kde = False);       # 直方图    密度估计
```

```python
sns.distplot(x,bins = 20,kde = False)     # 分成20个bins
```

```python
sns.distplot(x,bins = 20,kde = False,rug=True)
```

```python
from scipy import stats
```

```python
sns.distplot(x,bins = 20,kde = False,fit = stats.gamma)   # fit 控制拟合的分布图形
```

```python
# sns.kdeplot(x)       
sns.distplot(x,hist=False)       # the same as
```

```python
import pandas as pd
```

```python
mean ,cov = [0,1],[(1,.5),(.5,1)]
data = np.random.multivariate_normal(mean,cov,200)
df = pd.DataFrame(data,columns = ["X","Y"])
df.head()    # 共200个 只显示前5个  200 X 2
```

```python
sns.jointplot(x = "X",y="Y",data =df)     # 显示双变量关系的图
```

```python
sns.jointplot(x = "X",y="Y",data =df,color=sns.xkcd_rgb["pale red"],edgecolor = sns.xkcd_rgb["pale green"]) # 不知道如何两个变量不同颜色
```

```python
x,y = np.random.multivariate_normal(mean,cov,1000).T
with sns.axes_style("white"):
    sns.jointplot(x = x,y= y,kind = "hex",color = "k")     # 颜色越深点出现次数越多
    # kind={scatter,hex,reg,resid,kde}   hex最好看
```
