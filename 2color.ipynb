```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
%matplotlib inline
```

```python
sns.set(rc = {"figure.figsize":(6,6)})
```

```python
current_palette = sns.color_palette()  #默认颜色  怎么这么多
sns.palplot(current_palette)
```

```python
sns.palplot(sns.color_palette("hls",20))  # 默认颜色空间分为20份
```

```python
data = np.random.normal(size = (20,8)) + np.arange(8)/2
sns.boxplot(data = data ,palette=sns.color_palette('hls',8))
```

```python
colors = ["windows blue", "amber", "greyish", "faded green", "dusty purple"]  # 自己规定颜色
sns.palplot(sns.xkcd_palette(colors))
```

```python
sns.palplot(sns.hls_palette(10,l=.7,s=.9)) # l 是亮度  s 是饱和度 属于[0,1]
```

```python
sns.palplot(sns.color_palette("Paired",9))  # 颜色成对出现 还是可以是奇数的 
```

```python
plt.plot([0,1],[0,1],sns.xkcd_rgb["pale red"],lw = 3)    # xkcd 中颜色的命名有好多
plt.plot([0,1],[0,2],sns.xkcd_rgb["medium green"],lw = 3)
plt.plot([0,1],[0,3],sns.xkcd_rgb["denim blue"],lw = 3)
```

```python
sns.palplot(sns.light_palette("Blue"))  
```

```python
sns.palplot(sns.light_palette("Blue",reverse = True))  # 反转
```

```python
sns.palplot(sns.dark_palette("Blue"))
```

```python
sns.palplot(sns.color_palette("cubehelix",8))  # 色调线性变换？？？
```

```python
sns.palplot(sns.cubehelix_palette(8,start = 5,rot = .75))  # 区域？
```

```python
x, y = np.random.multivariate_normal([0, 0], [[1, -.5], [-.5, 1]], size=300).T
pal = sns.dark_palette("red", as_cmap=True)
sns.kdeplot(x, y, cmap=pal)                            # 渐变红色
```

```python
sns.palplot(sns.light_palette((210, 90, 60), input="husl"))  # 指定颜色空间 画固定值的颜色
```
